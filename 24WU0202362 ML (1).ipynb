```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
df=pd.read_excel(r"C:\Users\Sowmya S\OneDrive\Desktop\air cargo supplychain.xlsx")
df
```

```python
df.head()
```

```python
df1=df.drop(columns=["Shipment ID"])
df1
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
binary_encoder = LabelEncoder()
df1["Origin Airport"] = binary_encoder.fit_transform(df1["Origin Airport"])
df1["Destination Airport"] = binary_encoder.fit_transform(df1["Destination Airport"])
df1["Mode of Transport"] = binary_encoder.fit_transform(df1["Mode of Transport"])
df1["Carrier Name"] = binary_encoder.fit_transform(df1["Carrier Name"])
df1["Weather Conditions"] = binary_encoder.fit_transform(df1["Weather Conditions"])
df1["Day of Week"] = binary_encoder.fit_transform(df1["Day of Week"])
df1["Delay Status"] = binary_encoder.fit_transform(df1["Delay Status"])
df1["Temperature-Sensitive"] = binary_encoder.fit_transform(df1["Temperature-Sensitive"])
df1["Perishable Goods"] = binary_encoder.fit_transform(df1["Perishable Goods"])
df1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = df1.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
X = df1.drop(columns=["Delay Status"])
y = df1["Delay Status"]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.20, random_state=20)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
accuracy_score = log_reg.score(X_test, y_test)
accuracy_score
```

```python
dt_clf = DecisionTreeClassifier(random_state=20)
dt_clf.fit(X_train, y_train)
```

```python
df1_score = dt_clf.score(X_test, y_test)
df1_score
```

```python
rf_clf = RandomForestClassifier(n_estimators=10, random_state=20)
rf_clf.fit(X_train, y_train)
```

```python
rf_clf_score=rf_clf.score(X_test,y_test)
rf_clf_score
```
